**IMPORT LIBRARIES**

```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
from google.colab.patches import cv2_imshow #used to show image since cv2.imshow isnt applicable in google colab
```

```python
source = '/content/butterfly.jpg' #ill be using butterfly image. The next step is to make it grayscale
image=cv2.imread(source,0)#make it grayscale
```

```python
cv2_imshow(image)
```

# **Canny Edge Detector**

```python
#apply gaussian blur
#blurring is need because edge detection is sensitive noise which cause smoothing
blur =cv2.GaussianBlur(image, (5,5), 0) #5X5 kernel
```

```python
#the image is blurred already
cv2_imshow(blur)
```

```python
#canny

#since we're using OpenCV, Canny can easily be implemented using cv2.Canny
#inside canny, non-maximum supression and hysteresis threshold is included already

canny = cv2.Canny(image, 100, 200) #adjust the upper to (200) and lower limit to (100)

#display the result
cv2_imshow(canny)
```

# **Log**

```python
log = cv2.Laplacian(blur, cv2.CV_64F, ksize=5)
cv2_imshow(log)
```

# **Analysis**

Both algorithm managed to detect an image's edge with good result. But, in between those two results, *one algorithm performed better* than the other, which is *Canny edge detector*. Canny edge detector performed better mainly because it is *way more complex* than log. Unlike log, canny does non-maximum supression which *gets rid of noise produced by all sorts of objects* *and color gradients* in an image. Other than that, it also includes a step to *discern between edge directions*. Log output noisier data because it is based on second derivative which means sensitive to noise while Canny is based on the first derivative which means *its not sensitive to noise* . ***Overall, canny edge detector is the more effective / better method*** because its ***better at reducing noisy data and detecting edges*** (object detection)
