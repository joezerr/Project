**Clahe**

*Clahe is used to improve the conrast of an image. It works by taking a small tiles in an image. The neighbooring tiles are combined using bilinear interpolation to remove boundaries*

```python
clahe = cv2.createCLAHE(clipLimit = 5)
output_image = clahe.apply(input_image) 

cv2_imshow(output_image)
```

**Comparison**

```python
#reference vs output

figure = plt.figure(figsize=(16, 8))

ax1 = figure.add_subplot(221)
ax1.title.set_text('Reference image')
ax1.imshow(reference_image)

ax2 = figure.add_subplot(222)
ax2.title.set_text('Output image')
ax2.imshow(output_image, cmap='gray')

ax3 = figure.add_subplot(223)
ax3.title.set_text('reference histogram')
ax3.bar(range(256),
        cv2.calcHist([reference_image],[0],None,[256],[0,256]).ravel())

ax4 = figure.add_subplot(224)
ax4.title.set_text('Output histogram')
ax4.bar(range(256),
        cv2.calcHist([output_image],[0],None,[256],[0,256]).ravel())
```
