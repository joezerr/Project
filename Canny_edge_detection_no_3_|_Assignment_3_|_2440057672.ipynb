*Learning experience*: Edge detection using Canny Edge Detector is considered the best among other edge detectors. It has 4 different stages, unlike other edge detectors like sobel and laplacian. The additional stages are supression of false edges using non-maximum supression and Hysteresis Thresholding. Non-maximum supression is used to thin out any thick  edges while hysteresis thresholding is used to determine the relevancy of a gradient whether it can be an edge or not.

IMPORT LIBRARIES

```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
from google.colab.patches import cv2_imshow #used to show image since cv2.imshow isnt applicable in google colab
from skimage import io #used to import image from the internet
```

```python
source = 'https://asset-a.grid.id/crop/0x0:0x0/x/photo/2022/07/15/panda-3875289_640jpg-20220715103626.jpg' #ill be using panda image. The next step is to make it grayscale
image = io.imread(source)[:,:,0] # gray-scale image
```

```python
cv2_imshow(image)
```

```python
#apply gaussian blur
#blurring is need because edge detection is highly sensitive noise which cause smoothing
blur =cv2.GaussianBlur(image, (5,5), 0) #5X5 kernel
```

```python
#the image is blurred already
cv2_imshow(blur)
```

```python
#canny

#since we're using OpenCV, Canny can easily be implemented using cv2.Canny
#inside canny, non-maximum supression and hysteresis threshold is included already

canny = cv2.Canny(blur, 75, 125) #adjust the upper to (125) and lower limit to (75)

#display the result
cv2_imshow(canny)
```

```python
#plot
plot_image = np.concatenate((blur, np.abs(canny)),axis=1)

plt.imshow(plot_image, "gray"), plt.title("Canny")
plt.show()
```
